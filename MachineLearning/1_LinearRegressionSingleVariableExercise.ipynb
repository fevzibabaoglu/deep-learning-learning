```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('../Data/canada_per_capita_income.csv', names=['year', 'pci'], skiprows=[0])
df.head(5)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['pci'])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(df['year'], df['pci'], color='red')
plt.plot(df['year'], reg.predict(df[['year']]))
```
