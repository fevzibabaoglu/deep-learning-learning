```python
import pandas as pd
df = pd.read_csv('../Data/heart.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df = df[df['RestingBP'] != 0]
df = df[df['Cholesterol'] != 0]
df.describe()
```

```python
zscore_limit = 3
df = df[abs((df['Age'] - df['Age'].mean()) / df['Age'].std()) <= zscore_limit]
df = df[abs((df['RestingBP'] - df['RestingBP'].mean()) / df['RestingBP'].std()) <= zscore_limit]
df = df[abs((df['Cholesterol'] - df['Cholesterol'].mean()) / df['Cholesterol'].std()) <= zscore_limit]
df = df[abs((df['MaxHR'] - df['MaxHR'].mean()) / df['MaxHR'].std()) <= zscore_limit]
df = df[abs((df['Oldpeak'] - df['Oldpeak'].mean()) / df['Oldpeak'].std()) <= zscore_limit]
df.head()
```

```python
df.shape
```

```python
print(df['Sex'].unique())
print(df['ChestPainType'].unique())
print(df['RestingECG'].unique())
print(df['ExerciseAngina'].unique())
print(df['ST_Slope'].unique())
```

```python
pd.set_option('future.no_silent_downcasting', True)

df['Sex'] = df['Sex'].replace({
    'M': 0,
    'F': 1
})

df['ChestPainType'] = df['ChestPainType'].replace({
    'ATA': 0,
    'NAP': 1,
    'ASY': 2,
    'TA': 3
})

df['RestingECG'] = df['RestingECG'].replace({
    'Normal': 0,
    'ST': 1,
    'LVH': 2
})

df['ExerciseAngina'] = df['ExerciseAngina'].replace({
    'N': 0,
    'Y': 1
})

df['ST_Slope'] = df['ST_Slope'].replace({
    'Up': 0,
    'Flat': 1,
    'Down': 2
})

df.head()
```

```python
X = df.drop(['HeartDisease'], axis='columns')
y = df['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(), X_scaled, y, cv=5).mean()
```

```python
model_bagging = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True
)
cross_val_score(model_bagging, X_scaled, y, cv=5).mean()
```

```python
cross_val_score(DecisionTreeClassifier(), X_scaled, y, cv=5).mean()
```

```python
model_bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True
)
cross_val_score(model_bagging, X_scaled, y, cv=5).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
cross_val_score(RandomForestClassifier(), X_scaled, y, cv=5).mean()
```
