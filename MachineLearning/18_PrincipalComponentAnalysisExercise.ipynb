```python
import pandas as pd
df = pd.read_csv('../Data/heart.csv')
df.head()
```

```python
df.describe()
```

```python
df.nunique()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df.shape
```

```python
zscore_limit = 3
df = df[abs((df['Age'] - df['Age'].mean()) / df['Age'].std()) <= zscore_limit]
df = df[abs((df['RestingBP'] - df['RestingBP'].mean()) / df['RestingBP'].std()) <= zscore_limit]
df = df[abs((df['Cholesterol'] - df['Cholesterol'].mean()) / df['Cholesterol'].std()) <= zscore_limit]
df = df[abs((df['MaxHR'] - df['MaxHR'].mean()) / df['MaxHR'].std()) <= zscore_limit]
df = df[abs((df['Oldpeak'] - df['Oldpeak'].mean()) / df['Oldpeak'].std()) <= zscore_limit]
df.head()
```

```python
df.shape
```

```python
X = df.drop(['HeartDisease'], axis='columns')
y = df['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params = {
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear'),
        'params': {
            'C': [1, 5, 10]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy']
        }
    },
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    }
}
```

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV

scores = []
for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_scaled, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
df_scores = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_scores
```

```python
from sklearn.decomposition import PCA
pca = PCA(4)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
scores = []
for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_pca, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
df_scores = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_scores
```
