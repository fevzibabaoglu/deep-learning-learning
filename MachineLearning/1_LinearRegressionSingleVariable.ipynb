```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('../Data/homeprices_train.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df['area'], df['price'], color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df['price'])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df['area'], df['price'], color='red')
plt.plot(df['area'], reg.predict(df[['area']]))
```

```python
reg.predict(pd.DataFrame({
    'area': [3300]
}))
```

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
x = 3300
y = m * x + b
y
```

```python
d = pd.read_csv('../Data/homeprices_predict.csv')
d.head(3)
```

```python
p = reg.predict(d)
p
```

```python
d['prices'] = p
d
```

```python
d.to_csv('../Data/homeprices_predict_result.csv', index=False)
```
